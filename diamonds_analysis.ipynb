```python
import pandas as pd
import matplotlib.pyplot as plt

# URL of the diamonds.csv file
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv"

# Loading the dataset into a Pandas DataFrame
df = pd.read_csv(url)

# Displaying the first 7 rows of the DataFrame
print("First 7 rows of the diamonds dataset:")
print(df.head(7))
print("\n")

# Task 1: Descriptive Statistics for Numerical Columns
print("Descriptive statistics for numerical columns:")
print(df.describe())
print("\n")

# Task 2: Descriptive Statistics for Categorical Columns
categorical_columns = ['cut', 'color', 'clarity']
for col in categorical_columns:
    print(f"Descriptive statistics for '{col}' column:")
    print(df[col].describe())
    print("\n")

# Task 3: Unique Values in Categorical Columns
for col in categorical_columns:
    print(f"Unique values in '{col}' column:")
    print(df[col].unique())
    print("\n")

# Task 4: Data Visualization - Histograms for Numerical Columns
numerical_columns = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

for col in numerical_columns:
    plt.figure(figsize=(8, 5))
    df[col].hist(bins=30, color='blue', alpha=0.7)
    plt.title(f"Histogram for {col.capitalize()}")
    plt.xlabel(col.capitalize())
    plt.ylabel('Frequency')
    plt.grid(False)
    plt.show()
```
